<h1 style="text-align: center">Apprentissage machine: le buzz!</h1><br>



- Intelligence artificielle, apprentissage machine (automatique ou statistique) et apprentissage profond (rseaux de neurones profonds) ne sont **pas** des synonymes.

<div></div><br>



- Boom de l'intelligence artificielle d aux performances de l'apprentissage profond (prix Turing 2018 pour Bengio, Hinton et LeCun), permises par l'explosion des gros jeux de donnes (*Big data*) et de la puissance computationnelle (GPU).

<div></div><br>



- Cependant, la pertinence de l'apprentissage machine n'est **pas** unique  des donnes massives!
