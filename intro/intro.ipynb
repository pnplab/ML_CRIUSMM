<h1 style="text-align: center">Apprentissage machine: le buzz!</h1><br>



- Intelligence artificielle, apprentissage machine (automatique ou statistique) et apprentissage profond (réseaux de neurones profonds) ne sont **pas** des synonymes.

<div></div><br>



- Boom de l'intelligence artificielle dû aux performances de l'apprentissage profond (prix Turing 2018 pour Bengio, Hinton et LeCun), permises par l'explosion des gros jeux de données (*Big data*) et de la puissance computationnelle (GPU).

<div></div><br>



- Cependant, la pertinence de l'apprentissage machine n'est **pas** unique à des données massives!





<div style="text-align: center;">

    

</div>

<h1 style="text-align: center">Glorification?</h1>



<div style="text-align: center;">

    

</div>

- Oui et non. L'apprentissage machine partage certaines techniques avec les statistiques inférentielles classiques.

<div></div><br>



- Mais l'usage de ces techniques et, surtout, leur finalité diffèrent de façon importante.



<div style="text-align: center;">

    

</div>




<h1 style="text-align: center">Importance de l'apprentissage machine pour la psychiatrie</h1>

<div></div><br>



- **_Psychiatrie de précision:_** Apprentissage machine supervisé pour des prédictions individuelles dans des perspectives diagnostique, pronostique et théranostique.

<div></div><br>



- **_Hétérogénéité des troubles psychiatriques:_** Apprentissage machine non supervisé dans une perspective transdiagnostique.

<h1 style="text-align: center">Plan de cours</h1>



<h3>Cours théoriques et démonstrations</h3>



- **_Cours 1:_** Intro à l'apprentissage machine

    - Stats inférentielles vs apprentissage machine

    - Apprentissage machine supervisé vs non supervisé

<div></div><br>



- **_Cours 2:_** Apprentissage machine supervisé - régression

    - Hyperparamètres et régularisation (ridge, lasso, elastic-net)

    - Méthodes analytique vs itérative

    - Validation (croisée, croisée nichée)

<div></div><br>



- **_Cours 3:_** Apprentissage machine supervisé - classification

    - Évaluation: matrice de confusion et courbe "ROC"

    - Régression logistique

    - Machines à vecteurs de support

<div></div><br>



- **_Cours 4:_** Apprentissage machine supervisé - classification (suite)

    - Arbres de décision

    - Bagging: forêts aléatoires

    - Boosting de gradient

    - Classification naïve bayésienne

    - K plus proches voisins

    - Sélection de caractéristiques

    - Classification multi-classes

<div></div><br>



- **_Cours 5:_** Apprentissage machine non supervisé

    - Réduction de dimensions (PCA, t-SNE)

    - Regroupement: k-moyennes, décalage de moyenne, DBSCAN, Mélange gaussien, Hiérarchique

    - Validation interne vs externe

<div></div><br>



- **_Cours 5:_** Réseaux de neurones (profonds)
